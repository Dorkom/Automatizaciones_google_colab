````python
"""
================================================================================
  MULTI-CONVERTER: PDF + DOCX → Markdown  |  Google Colab Edition  v1.0
  Autor  : Senior AI Engineer
  Propósito: Comparar 4 librerías de conversión para pipelines RAG.

  CONVERSORES INCLUIDOS:
    1. markitdown  — Microsoft. Generalista, rápido, PDF + DOCX nativo.
    2. pymupdf4llm — Artífex/PyMuPDF. El más preciso en PDFs con texto
                     seleccionable. DOCX vía mammoth.
    3. docling      — IBM Research. Orientado a RAG estructurado, excelente
                     con LlamaIndex / LangChain.
    4. marker       — Datalab. Mayor calidad de layout (usa ML). Más lento
                     pero el mejor en documentos complejos / multi-columna.

  SALIDA:
    /content/markdowns/
        markitdown/      ← .md generados por MarkItDown
        pymupdf4llm/     ← .md generados por PyMuPDF4LLM + mammoth
        docling/         ← .md generados por Docling
        marker/          ← .md generados por Marker
        COMPARACION.md   ← tabla resumen con chars, tiempo y errores
    /content/markdowns_comparacion.zip   ← todo empaquetado para descargar

  CELDA DE INSTALACIÓN (ejecutar primero):
    !pip install -q markitdown[all] pypdf python-docx
    !pip install -q pymupdf4llm mammoth
    !pip install -q docling
    !pip install -q marker-pdf
================================================================================
"""

# ============================================================
# SECCIÓN 1 — IMPORTACIONES ESTÁNDAR
# ============================================================
import os
import re
import time
import zipfile
import logging
import unicodedata
import traceback
from pathlib import Path
from datetime import datetime

# ============================================================
# SECCIÓN 2 — CONFIGURACIÓN GLOBAL
# ============================================================

INPUT_DIR    = Path("/content/docs")           # PDFs y DOCXs de entrada
OUTPUT_DIR   = Path("/content/markdowns")      # Raíz de salida (subcarpetas por conversor)
ZIP_OUTPUT   = Path("/content/markdowns_comparacion.zip")
INPUT_EXTENSIONS: tuple[str, ...] = (".pdf", ".docx")

# Metadatos de cada conversor (nombre amigable, descripción breve)
CONVERTER_META: dict[str, str] = {
    "markitdown":  "Microsoft MarkItDown — generalista, PDF + DOCX nativo",
    "pymupdf4llm": "PyMuPDF4LLM (PDF) + mammoth (DOCX) — mejor en PDFs con texto",
    "docling":     "IBM Docling — ideal para RAG / LlamaIndex / LangChain",
    "marker":      "Datalab Marker — mayor calidad de layout, usa ML",
}

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s  [%(levelname)s]  %(message)s",
    datefmt="%H:%M:%S",
)
log = logging.getLogger(__name__)


# ============================================================
# SECCIÓN 3 — PATRONES REGEX DE LIMPIEZA
# (idénticos al pipeline v3; se aplican a TODOS los conversores)
# ============================================================

HEADER_FOOTER_PATTERNS = [
    re.compile(r"(?im)^universidad\s+de\s+lima\s*[|\-–]\s*\d+\s*$"),
    re.compile(r"(?im)^.{5,60}[–\-]\s*p[aá]g\.?\s*\d+\s*$"),
    re.compile(r"(?m)^\s*\d{1,3}\s*$"),
    re.compile(r"(?im)^reglamento\s+(general|de\s+\w+)\s*$"),
]

TITULO_PATTERN = re.compile(
    r"(?im)^(t[íi]tulo\s+[IVXLCDM\d]+[.:)]*\s*.{0,80})$"
)
CAPITULO_PATTERN = re.compile(
    r"(?im)^(cap[íi]tulo\s+(?:[IVXLCDM\d]+|primero|segundo|tercero|cuarto|"
    r"quinto|sexto|s[eé]ptimo|octavo|noveno|d[eé]cimo)[.:)]*\s*.{0,80})$"
)
ARTICULO_PATTERN = re.compile(
    r"(?im)^(art[íi]?culo\.?\s*\d+[°º]?\.?-?\s*.{0,120})$"
)
INDICE_START_PATTERN = re.compile(r"(?im)^\s*[íi]ndice\s*$")
INDICE_END_PATTERN   = re.compile(r"(?m)^#{1,2}\s+\S")


# ============================================================
# SECCIÓN 4 — PIPELINE DE LIMPIEZA
# (se reutiliza en los 4 conversores para comparar solo la
#  calidad de extracción, no el post-procesado)
# ============================================================

def normalize_unicode(text: str) -> str:
    text = unicodedata.normalize("NFC", text)
    text = text.replace("\u201c", '"').replace("\u201d", '"')
    text = text.replace("\u2018", "'").replace("\u2019", "'")
    text = text.replace("\u2013", "-").replace("\u2014", "-")
    return text


def remove_headers_footers(text: str) -> str:
    lines = text.splitlines()
    return "\n".join(
        line for line in lines
        if not any(pat.match(line.strip()) for pat in HEADER_FOOTER_PATTERNS)
    )


def _identify_index_line_range(lines: list[str]) -> tuple[int, int]:
    start_idx = next(
        (i for i, l in enumerate(lines) if INDICE_START_PATTERN.match(l.strip())),
        -1,
    )
    if start_idx == -1:
        return (-1, -1)
    end_idx = next(
        (j for j in range(start_idx + 1, len(lines)) if INDICE_END_PATTERN.match(lines[j])),
        len(lines),
    )
    return (start_idx, end_idx)


def normalize_headings(text: str) -> str:
    lines = text.splitlines()
    index_start, index_end = _identify_index_line_range(lines)
    index_range = range(index_start, index_end) if index_start != -1 else range(0, 0)

    def _prefix(line: str, hashes: str) -> str:
        stripped = line.strip()
        return line if stripped.startswith("#") else f"{hashes} {stripped}"

    result = []
    for i, line in enumerate(lines):
        if i in index_range or "|" in line:
            result.append(line)
            continue
        stripped = line.strip()
        if TITULO_PATTERN.match(stripped):
            result.append(_prefix(stripped, "#"))
        elif CAPITULO_PATTERN.match(stripped):
            result.append(_prefix(stripped, "##"))
        elif ARTICULO_PATTERN.match(stripped):
            result.append(_prefix(stripped, "###"))
        else:
            result.append(line)
    return "\n".join(result)


def fix_orphan_line_breaks(text: str) -> str:
    """Une saltos de línea huérfanos de PDFs (no aplica a DOCX)."""
    lines = text.splitlines()
    result = []
    i = 0
    while i < len(lines):
        current = lines[i]
        is_structural = (
            current.strip().startswith(("#", "|", "-", "*", "+", ">"))
            or re.match(r"^\s*\d+\.\s", current)
        )
        if (
            not is_structural
            and i + 1 < len(lines)
            and current.strip()
            and not re.search(r"[.;:?!\|]\s*$", current)
        ):
            next_line = lines[i + 1]
            next_is_structural = (
                next_line.strip().startswith(("#", "|", "-", "*", "+", ">"))
                or re.match(r"^\s*\d+\.\s", next_line)
                or next_line.strip() == ""
            )
            if not next_is_structural and next_line.strip():
                result.append(current + " " + next_line.strip())
                i += 2
                continue
        result.append(current)
        i += 1
    return "\n".join(result)


def fix_docx_bold_headings(text: str) -> str:
    """Corrige encabezados envueltos en negrita que genera markitdown con DOCX."""
    BOLD_HEADING = re.compile(
        r"^\*{1,2}\s*(#{1,3}\s+.+?)\s*\*{1,2}\s*$",
        re.MULTILINE,
    )
    return BOLD_HEADING.sub(r"\1", text)


def remove_excessive_blank_lines(text: str) -> str:
    return re.sub(r"\n{3,}", "\n\n", text)


def clean_markdown(raw_md: str, source_format: str = "pdf") -> str:
    """
    Pipeline de limpieza completo. Igual para todos los conversores.
    source_format: 'pdf' o 'docx'
    """
    text = normalize_unicode(raw_md)
    text = remove_headers_footers(text)
    text = normalize_headings(text)           # ANTES de fix_orphan (BUG FIX #2)
    if source_format == "pdf":
        text = fix_orphan_line_breaks(text)
    else:
        text = fix_docx_bold_headings(text)
    text = remove_excessive_blank_lines(text)
    return text.strip()


# ============================================================
# SECCIÓN 5 — ADAPTADORES DE CONVERSIÓN
# Cada función recibe un Path y devuelve el Markdown CRUDO (str).
# El pipeline de limpieza se aplica después, de forma centralizada.
# ============================================================

# ── Cache global de modelos pesados (evita recargar en cada archivo) ──────────
_marker_models_cache = None


def convert_markitdown(file_path: Path) -> str:
    """
    Conversor 1: Microsoft MarkItDown
    Soporta PDF y DOCX de forma nativa. Muy rápido y confiable.
    """
    from markitdown import MarkItDown
    result = MarkItDown().convert(str(file_path))
    raw = result.text_content
    if not raw or len(raw.strip()) < 50:
        raise RuntimeError("MarkItDown devolvió contenido vacío o insuficiente.")
    return raw


def convert_pymupdf4llm(file_path: Path) -> str:
    """
    Conversor 2: PyMuPDF4LLM (PDF) + mammoth (DOCX)
    PyMuPDF4LLM: excelente para PDFs nativos, extrae tablas y estructura.
    mammoth: convierte DOCX preservando la jerarquía de estilos de Word.
    """
    ext = file_path.suffix.lower()
    if ext == ".pdf":
        import pymupdf4llm
        raw = pymupdf4llm.to_markdown(
            str(file_path),
            # Desactiva encabezados/pies de página detectados automáticamente
            # para que nuestro pipeline de limpieza los maneje uniformemente.
            # Nota: parámetros disponibles según versión instalada.
        )
    elif ext == ".docx":
        import mammoth
        with open(file_path, "rb") as f:
            result = mammoth.convert_to_markdown(f)
        raw = result.value
        # mammoth incluye advertencias útiles en result.messages
        if result.messages:
            for msg in result.messages[:3]:  # log las 3 primeras advertencias
                log.debug("    mammoth warning: %s", msg)
    else:
        raise ValueError(f"Extensión no soportada por pymupdf4llm+mammoth: {ext}")

    if not raw or len(raw.strip()) < 50:
        raise RuntimeError("pymupdf4llm / mammoth devolvió contenido vacío.")
    return raw


def convert_docling(file_path: Path) -> str:
    """
    Conversor 3: IBM Docling
    Soporte nativo de PDF y DOCX. Muy buena integración con LlamaIndex
    y LangChain. Detecta automáticamente tablas, listas y jerarquías.
    """
    from docling.document_converter import DocumentConverter
    converter = DocumentConverter()
    result = converter.convert(str(file_path))
    raw = result.document.export_to_markdown()
    if not raw or len(raw.strip()) < 50:
        raise RuntimeError("Docling devolvió contenido vacío.")
    return raw


def convert_marker(file_path: Path) -> str:
    """
    Conversor 4: Datalab Marker
    El de mayor calidad en documentos complejos: multi-columna, tablas
    anidadas, fórmulas, multi-idioma. Usa modelos ML (torch).
    Carga los modelos una sola vez y los reutiliza entre archivos.

    Compatible con marker-pdf >= 0.3 (API antigua) y >= 1.0 (API nueva).
    Para DOCX en versiones sin DocxConverter, recae en mammoth.
    """
    global _marker_models_cache
    ext = file_path.suffix.lower()

    # ── Intentar API moderna (marker >= 1.0) ──────────────────────────────────
    try:
        from marker.config.parser import ConfigParser
        from marker.models import create_model_dict

        if _marker_models_cache is None:
            log.info("  🧠  Cargando modelos de Marker (primera vez, puede tardar 1-2 min)...")
            _marker_models_cache = create_model_dict()
            log.info("  🧠  Modelos de Marker listos y en caché.")

        config_parser = ConfigParser({"output_format": "markdown"})
        cfg = config_parser.generate_config_dict()

        if ext == ".pdf":
            from marker.converters.pdf import PdfConverter
            converter = PdfConverter(config=cfg, artifact_dict=_marker_models_cache)
            rendered  = converter(str(file_path))
            raw       = rendered.markdown

        elif ext == ".docx":
            # Intentar conversor nativo de marker para DOCX
            try:
                from marker.converters.docx import DocxConverter
                converter = DocxConverter(config=cfg, artifact_dict=_marker_models_cache)
                rendered  = converter(str(file_path))
                raw       = rendered.markdown
            except ImportError:
                # Esta versión de marker no tiene DocxConverter → usar mammoth
                log.warning(
                    "  ⚠️  marker no incluye DocxConverter en esta versión. "
                    "Usando mammoth como fallback para DOCX."
                )
                import mammoth
                with open(file_path, "rb") as f:
                    result = mammoth.convert_to_markdown(f)
                raw = result.value
        else:
            raise ValueError(f"Extensión no soportada: {ext}")

    except ImportError:
        # ── Fallback: API antigua marker (< 1.0) — solo PDF ──────────────────
        if ext != ".pdf":
            raise RuntimeError(
                "Marker versión antigua solo soporta PDF. "
                "Instala marker-pdf >= 1.0 para soporte de DOCX."
            )
        log.warning("  ⚠️  Usando API antigua de marker (< 1.0).")
        from marker.convert import convert_single_pdf
        from marker.models import load_all_models

        if _marker_models_cache is None:
            log.info("  🧠  Cargando modelos de Marker (API antigua)...")
            _marker_models_cache = load_all_models()
        raw, _, _ = convert_single_pdf(str(file_path), _marker_models_cache)

    if not raw or len(raw.strip()) < 50:
        raise RuntimeError("Marker devolvió contenido vacío.")
    return raw


# Registro centralizado: clave → función de conversión
CONVERTERS: dict[str, callable] = {
    "markitdown":  convert_markitdown,
    "pymupdf4llm": convert_pymupdf4llm,
    "docling":     convert_docling,
    "marker":      convert_marker,
}


# ============================================================
# SECCIÓN 6 — GUARDADO CON YAML FRONT-MATTER
# ============================================================

def save_markdown(
    content: str,
    output_path: Path,
    source_ext: str,
    converter_name: str,
) -> None:
    """
    Escribe el Markdown en disco con YAML front-matter enriquecido.
    El campo 'converter' permite filtrar por librería en el pipeline RAG.
    """
    stem = output_path.stem
    fmt  = source_ext.lstrip(".")
    yaml = (
        "---\n"
        f"source: {stem}{source_ext}\n"
        f"title: {stem.replace('_', ' ').title()}\n"
        "type: reglamento_institucional\n"
        f"format: {fmt}\n"
        f"converter: {converter_name}\n"
        "---\n\n"
    )
    output_path.write_text(yaml + content, encoding="utf-8")


# ============================================================
# SECCIÓN 7 — REPORTE DE COMPARACIÓN
# ============================================================

def generate_comparison_report(stats: dict[str, dict]) -> str:
    """
    Genera un Markdown con:
      - Tabla resumen por conversor (exitosos, fallidos, tiempo total)
      - Tabla detallada por archivo × conversor
      - Guía de interpretación

    stats = {
        "markitdown": {
            "reglamento.pdf": {"status": "ok", "chars": 12345, "time_s": 1.8},
            "norma.docx":     {"status": "error", "error": "...", "time_s": 0.2},
        },
        ...
    }
    """
    now = datetime.now().strftime("%Y-%m-%d %H:%M")
    lines: list[str] = []

    lines += [
        "# 📊 Reporte de Comparación: PDF/DOCX → Markdown",
        "",
        f"> Generado el {now}",
        "",
        "Compara los `.md` de cada subcarpeta para elegir el conversor que mejor",
        "preserve la estructura de tus reglamentos antes de indexarlos en tu RAG.",
        "",
        "---",
        "",
        "## Resumen por Convertidor",
        "",
        "| Convertidor | Descripción | ✅ OK | ❌ Error | ⏭️ Omitido | ⏱️ Tiempo (s) |",
        "|---|---|:---:|:---:|:---:|---:|",
    ]

    for conv_name, file_results in stats.items():
        desc    = CONVERTER_META.get(conv_name, conv_name)
        ok      = sum(1 for r in file_results.values() if r["status"] == "ok")
        err     = sum(1 for r in file_results.values() if r["status"] == "error")
        skipped = sum(1 for r in file_results.values() if r["status"] == "omitido")
        t_total = sum(r.get("time_s", 0) for r in file_results.values())
        lines.append(
            f"| `{conv_name}` | {desc} | {ok} | {err} | {skipped} | {t_total:.1f} |"
        )

    lines += ["", "---", "", "## Detalle por Archivo", ""]

    all_files = sorted({f for fr in stats.values() for f in fr})
    for fname in all_files:
        lines += [
            f"### `{fname}`",
            "",
            "| Convertidor | Estado | Caracteres limpios | Tiempo (s) | Observación |",
            "|---|:---:|---:|---:|---|",
        ]
        for conv_name, file_results in stats.items():
            r      = file_results.get(fname, {"status": "no ejecutado"})
            status = r["status"]
            icon   = {"ok": "✅", "error": "❌", "omitido": "⏭️"}.get(status, "❓")
            chars  = f"{r['chars']:,}" if status == "ok" else "—"
            t_str  = f"{r['time_s']:.1f}" if status in ("ok", "error") else "—"
            note   = (r.get("error", "")[:90] + "…") if status == "error" else ""
            lines.append(
                f"| `{conv_name}` | {icon} {status} | {chars} | {t_str} | {note} |"
            )
        lines.append("")

    lines += [
        "---",
        "",
        "## 🔍 Guía de Interpretación",
        "",
        "| Convertidor | Mejor para | Limitaciones |",
        "|---|---|---|",
        "| `markitdown` | Uso general, formatos mixtos | Fidelidad media en PDFs complejos |",
        "| `pymupdf4llm` | PDFs con texto seleccionable, tablas | DOCX via mammoth (pérdida de estilos avanzados) |",
        "| `docling` | Pipelines RAG (LlamaIndex, LangChain) | Más lento; dependencias pesadas |",
        "| `marker` | Documentos complejos, multi-columna | El más lento; requiere GPU para máxima velocidad |",
        "",
        "**Recomendación para reglamentos institucionales:**",
        "1. Revisa visualmente 2-3 documentos con cada conversor.",
        "2. Prioriza el que mejor preserve los encabezados jerárquicos (TÍTULO / CAPÍTULO / ARTÍCULO).",
        "3. Verifica que las tablas del índice no se mezclen con el cuerpo.",
        "4. Usa `MarkdownHeaderTextSplitter` de LangChain como validación final.",
        "",
        "```python",
        "# Snippet de validación rápida",
        "from langchain.text_splitter import MarkdownHeaderTextSplitter",
        "",
        "HEADERS = [(\"#\", \"titulo\"), (\"##\", \"capitulo\"), (\"###\", \"articulo\")]",
        "splitter = MarkdownHeaderTextSplitter(headers_to_split_on=HEADERS)",
        "md = open('/content/markdowns/<conversor>/mi_reglamento.md').read()",
        "chunks = splitter.split_text(md)",
        "print(f'Chunks: {len(chunks)} | Primer chunk: {chunks[0].metadata}')",
        "```",
    ]

    return "\n".join(lines)


# ============================================================
# SECCIÓN 8 — COMPRESIÓN
# ============================================================

def zip_output_folder() -> None:
    """Empaqueta toda la carpeta OUTPUT_DIR (subcarpetas + reporte) en un ZIP."""
    all_files = list(OUTPUT_DIR.rglob("*.md"))
    if not all_files:
        log.warning("⚠️  No hay archivos .md para comprimir.")
        return

    with zipfile.ZipFile(ZIP_OUTPUT, "w", compression=zipfile.ZIP_DEFLATED) as zf:
        for f in all_files:
            # arcname relativo a OUTPUT_DIR para que el ZIP tenga estructura de carpetas
            zf.write(f, arcname=f.relative_to(OUTPUT_DIR))

    log.info(
        "📦  ZIP generado: %s  (%d archivos, %.1f KB)",
        ZIP_OUTPUT,
        len(all_files),
        ZIP_OUTPUT.stat().st_size / 1024,
    )


# ============================================================
# SECCIÓN 9 — ORQUESTADOR PRINCIPAL
# ============================================================

def process_batch(
    converters_to_run: list[str] | None = None,
    skip_on_error: bool = True,
) -> dict[str, dict]:
    """
    Orquesta el procesamiento completo del lote con múltiples conversores.

    Args:
        converters_to_run : Lista de nombres de conversores a ejecutar.
                            None = usar todos (CONVERTERS.keys()).
                            Ej: ["markitdown", "pymupdf4llm"]
        skip_on_error     : Si True (defecto), un fallo en un conversor/archivo
                            no detiene el resto del batch.

    Returns:
        dict con estadísticas por conversor → archivo → resultado.
    """
    INPUT_DIR.mkdir(parents=True, exist_ok=True)
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

    active = converters_to_run or list(CONVERTERS.keys())

    # Validar nombres de conversores
    unknown = [c for c in active if c not in CONVERTERS]
    if unknown:
        raise ValueError(f"Conversores desconocidos: {unknown}. Válidos: {list(CONVERTERS)}")

    # Crear subcarpetas de salida
    for conv_name in active:
        (OUTPUT_DIR / conv_name).mkdir(parents=True, exist_ok=True)

    # Recopilar archivos de entrada
    input_files = sorted(
        f for ext in INPUT_EXTENSIONS for f in INPUT_DIR.glob(f"*{ext}")
    )
    if not input_files:
        log.error("❌  No se encontraron archivos en '%s'. Sube tus PDFs/DOCXs.", INPUT_DIR)
        return {}

    stats: dict[str, dict] = {c: {} for c in active}
    total_tasks = len(input_files) * len(active)

    log.info("═" * 70)
    log.info("🚀  BATCH MULTI-CONVERSOR")
    log.info("   Archivos : %d  |  Conversores: %d  |  Tareas totales: %d",
             len(input_files), len(active), total_tasks)
    log.info("   Conversores activos: %s", ", ".join(active))
    log.info("═" * 70)

    task_num = 0
    for file_idx, file_path in enumerate(input_files, 1):
        ext = file_path.suffix.lower()
        fmt = ext.lstrip(".")
        icon = "📄" if fmt == "pdf" else "📝"
        log.info("\n%s  [Archivo %d/%d]  %s", icon, file_idx, len(input_files), file_path.name)

        for conv_name in active:
            task_num += 1
            conv_func   = CONVERTERS[conv_name]
            output_dir  = OUTPUT_DIR / conv_name
            output_path = output_dir / (file_path.stem + ".md")

            log.info("  ├─ [%d/%d]  %-14s  …", task_num, total_tasks, conv_name)

            # Modo incremental: omitir si el .md ya existe
            if output_path.exists():
                log.info("  │   ↩️  Ya existe, omitiendo.")
                stats[conv_name][file_path.name] = {"status": "omitido"}
                continue

            t0 = time.time()
            try:
                # A) Conversión cruda
                raw_md   = conv_func(file_path)
                raw_chars = len(raw_md)

                # B) Pipeline de limpieza (uniforme para todos)
                clean_md = clean_markdown(raw_md, source_format=fmt)

                # C) Guardado en disco
                save_markdown(clean_md, output_path, source_ext=ext, converter_name=conv_name)

                elapsed = time.time() - t0
                chars   = len(clean_md)
                pct     = 100 * (1 - chars / raw_chars) if raw_chars else 0

                log.info(
                    "  │   ✅  %d chars limpios  (reducción %.1f%%)  %.1fs",
                    chars, pct, elapsed,
                )
                stats[conv_name][file_path.name] = {
                    "status": "ok",
                    "chars":  chars,
                    "time_s": round(elapsed, 2),
                }

            except Exception as exc:
                elapsed = time.time() - t0
                err_msg = str(exc)
                log.error("  │   ❌  FALLO (%s): %s", conv_name, err_msg)
                if not skip_on_error:
                    raise
                stats[conv_name][file_path.name] = {
                    "status": "error",
                    "error":  err_msg,
                    "time_s": round(elapsed, 2),
                }

        log.info("  └─ ✔  Archivo completado.")

    # ── Reporte de comparación ─────────────────────────────────────────────────
    log.info("\n📝  Generando reporte de comparación...")
    report_md   = generate_comparison_report(stats)
    report_path = OUTPUT_DIR / "COMPARACION.md"
    report_path.write_text(report_md, encoding="utf-8")
    log.info("   Guardado: %s", report_path)

    # ── Comprimir ──────────────────────────────────────────────────────────────
    log.info("📦  Comprimiendo salida...")
    zip_output_folder()

    # ── Resumen final ──────────────────────────────────────────────────────────
    log.info("\n" + "═" * 70)
    log.info("📊  RESUMEN FINAL")
    log.info("%-16s  %6s  %6s  %7s  %s", "CONVERSOR", "OK", "ERROR", "OMIT", "T.TOTAL(s)")
    log.info("─" * 70)
    for conv_name, file_results in stats.items():
        ok   = sum(1 for r in file_results.values() if r["status"] == "ok")
        err  = sum(1 for r in file_results.values() if r["status"] == "error")
        skip = sum(1 for r in file_results.values() if r["status"] == "omitido")
        t    = sum(r.get("time_s", 0) for r in file_results.values())
        log.info("%-16s  %6d  %6d  %7d  %.1f", conv_name, ok, err, skip, t)
    log.info("═" * 70)
    log.info("🎉  Descarga el ZIP desde: %s", ZIP_OUTPUT)
    log.info("═" * 70)

    return stats

# ============================================================
# SECCIÓN 10 — PUNTO DE ENTRADA Y SELECCIÓN INTERACTIVA
# ============================================================

def prompt_converter_selection() -> list[str] | None:
    """
    Muestra un menú interactivo para elegir qué conversor ejecutar.
    Por defecto (si el usuario presiona Enter), devuelve ['marker'].
    Si elige 'Todos', devuelve None (que process_batch interpreta como todos).
    """
    print("\n" + "═" * 50)
    print(" 🛠️  SELECCIÓN DE CONVERSOR RAG")
    print("═" * 50)
    print("Selecciona el conversor que deseas evaluar:")
    print("  1. markitdown")
    print("  2. pymupdf4llm")
    print("  3. docling")
    print("  4. marker (POR DEFECTO)")
    print("  5. Ejecutar TODOS")
    print("─" * 50)

    opciones = {
        "1": ["markitdown"],
        "2": ["pymupdf4llm"],
        "3": ["docling"],
        "4": ["marker"],
        "5": None  # None le indica a process_batch que use list(CONVERTERS.keys())
    }

    while True:
        choice = input("Ingresa el número de tu elección [4]: ").strip()

        # Si presiona Enter sin escribir nada, retorna marker por defecto
        if not choice:
            return ["marker"]

        if choice in opciones:
            return opciones[choice]

        print("❌ Opción no válida. Por favor ingresa un número del 1 al 5, o presiona Enter para usar 'marker'.")

if __name__ == "__main__":
    # ──────────────────────────────────────────────────────────────────────────
    # INSTRUCCIONES DE USO EN GOOGLE COLAB:
    #
    #  ① INSTALACIÓN (Celda 1 — ejecutar una vez):
    #
    #    !pip install -q markitdown[all] pypdf python-docx
    #    !pip install -q pymupdf4llm mammoth
    #    !pip install -q docling
    #    !pip install -q marker-pdf
    #
    #  ② SUBIR ARCHIVOS a /content/docs/  (Celda 2):
    #
    #    from google.colab import files
    #    import shutil, os
    #    os.makedirs('/content/docs', exist_ok=True)
    #    uploaded = files.upload()
    #    for fname in uploaded:
    #        shutil.move(fname, f'/content/docs/{fname}')
    #
    #  ③ EJECUTAR este script (Celda 3):
    #
    #    # El script ahora te pedirá de forma interactiva qué conversor usar.
    #
    #  ④ DESCARGAR el ZIP (Celda 4):
    #
    #    from google.colab import files
    #    files.download('/content/markdowns_comparacion.zip')
    #
    # ─────────────────────────────────────────────────────────────────────────
    #  ESTRUCTURA DEL ZIP:
    #    markdowns_comparacion.zip
    #    ├── markitdown/         ← un .md por documento
    #    ├── pymupdf4llm/
    #    ├── docling/
    #    ├── marker/
    #    └── COMPARACION.md       ← tabla de stats: chars, tiempo, errores
    #
    #  NOTA SOBRE TIEMPOS ESPERADOS (por archivo, CPU Colab gratuito):
    #    markitdown  :  ~2-5 s
    #    pymupdf4llm :  ~3-8 s  (PDF)  |  ~1-2 s  (DOCX via mammoth)
    #    docling     :  ~10-30 s  (carga modelos en primer archivo)
    #    marker      :  ~30-120 s  (carga modelos ~2 min solo la primera vez)
    # ──────────────────────────────────────────────────────────────────────────

    # 1. Preguntar al usuario qué conversor desea usar
    conversores_seleccionados = prompt_converter_selection()

    # 2. Ejecutar el orquestador con la selección
    results = process_batch(converters_to_run=conversores_seleccionados)


# ============================================================
# SECCIÓN 11 — SNIPPET DE VALIDACIÓN CON LANGCHAIN (opcional)
# ============================================================
# Ejecuta esto en una celda separada para verificar la calidad
# del chunking semántico en cada conversor:
#
# from langchain.text_splitter import MarkdownHeaderTextSplitter
# from pathlib import Path
#
# HEADERS = [("#", "titulo"), ("##", "capitulo"), ("###", "articulo")]
# splitter = MarkdownHeaderTextSplitter(headers_to_split_on=HEADERS)
#
# OUTPUT_DIR = Path("/content/markdowns")
# for conv_dir in sorted(OUTPUT_DIR.iterdir()):
#     if not conv_dir.is_dir():
#         continue
#     md_files = list(conv_dir.glob("*.md"))
#     if not md_files:
#         continue
#     sample = md_files[0].read_text(encoding="utf-8")
#     chunks = splitter.split_text(sample)
#     print(f"\n{'='*50}")
#     print(f"  Conversor : {conv_dir.name}")
#     print(f"  Archivo   : {md_files[0].name}")
#     print(f"  Chunks    : {len(chunks)}")
#     if chunks:
#         print(f"  Primer chunk metadata : {chunks[0].metadata}")
#         print(f"  Primer chunk preview  : {chunks[0].page_content[:150]!r}")
````
